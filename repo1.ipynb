```python
model = Model("M")

#å¤‰æ•°å®šç¾©
x11 = model.addVar(name="x11")
x12 = model.addVar(name="x12")
x21 = model.addVar(name="x21")
x22 = model.addVar(name="x22")
y11 = model.addVar(name="y11")
y12 = model.addVar(name="y12")
y13 = model.addVar(name="y13")
y14 = model.addVar(name="y14")
y21 = model.addVar(name="y21")
y22 = model.addVar(name="y22")
y23 = model.addVar(name="y23")
y24 = model.addVar(name="y24")
z1 = model.addVar(name="z1")
z2 = model.addVar(name="z2")

#å¿˜ã‚Œãšã«
model.update()  

#åˆ¶ç´„

#â‘ å¤‰æ•°ã®éè² æ¡ä»¶ï¼ˆ14å¤‰æ•°ï¼‰
model.addConstr(x11 >= 0)
model.addConstr(x12 >= 0)
model.addConstr(x21 >= 0)
model.addConstr(x22 >= 0)
model.addConstr(y11 >= 0)
model.addConstr(y12 >= 0)
model.addConstr(y13 >= 0)
model.addConstr(y14 >= 0)
model.addConstr(y21 >= 0)
model.addConstr(y22 >= 0)
model.addConstr(y23 >= 0)
model.addConstr(y24 >= 0)
model.addConstr(z1 >= 0)
model.addConstr(z2 >= 0)

#â‘¡ãƒ•ãƒ­ãƒ¼ä¿å­˜å‰‡ï¼ˆé…é€ã‚»ãƒ³ã‚¿ãƒ¼ã«æ¬å…¥ã•ã‚Œã‚‹ãƒˆãƒ©ãƒƒã‚¯å®¹é‡ã¯ï¼Œé…é€ã‚»ãƒ³ã‚¿ãƒ¼ã‹ã‚‰ã®æ¬å‡ºå®¹é‡ä»¥ä¸‹ã¨ãªã‚‹ï¼ã§ãªã„ã¨é…é€ã‚»ãƒ³ã‚¿ãƒ¼ã«åœ¨åº«ãŒãŸã¾ã£ã¦ã—ã¾ã†ï¼‰
model.addConstr(100 * x11 +100 * x21 -40 * y11 -40 * y12  -40 * y13 
                -40 * y14<=0)
model.addConstr(100 * x12 +100 * x22 -40 * y21 -40 * y22 -40 * y23  -40 * y24<=0)

#â‘¢ãƒˆãƒ©ãƒƒã‚¯é›†ä¸­å®¹é‡ï¼ˆÎ£ğ‘¦ğ‘–ğ‘—2ğ‘–=1ãªã©ï¼‰ã¯ï¼Œå„éƒ½çœŒã®éœ€è¦ã‚’ä¸Šå›ã‚‹
model.addConstr(40 * y11 +40 * y21>= 500)
model.addConstr(40 * y12 +40 * y22>= 300)
model.addConstr(40 * y13 +40 * y23>= 200)
model.addConstr(40 * y14 +40 * y24>= 200)

#â‘£10tãƒˆãƒ©ãƒƒã‚¯ã®å®¹é‡åˆè¨ˆ Î£Î£ğ‘¥ğ‘–ğ‘—2ğ‘—=12ğ‘–=1 ãŒï¼Œå·¥å ´ã‹ã‚‰ã®ç™ºç”Ÿé‡ã‚’ä¸Šå›ã‚‹
model.addConstr(100 * x11 +100 * x12+100 * x21 +100 * x22>=1200)

#â‘¤æ–°è¦æ–½è¨­é¢ç©ã¨æ—¢å­˜é¢ç©ã®å’Œã¯æ¬å…¥ã•ã‚Œã‚‹ã‚±ãƒ¼ã‚¹æ•°ã«æ¯”ä¾‹ã—ãŸå¿…è¦é¢ç©ã‚’ä¸Šå›ã‚‹å¿…è¦ãŒã‚ã‚‹
model.addConstr(5000 +  z1 >= 5000/3 * x11 +5000/3 * x21)
model.addConstr(5000 +  z2 >= 5000/3 * x12 +5000/3 * x22)

#ç›®çš„é–¢æ•°
model.setObjective(6000 * x11 + 8000 * x12 + 8000 * x21+ 6000 * x22+ 1600 * y11+ 3000 * y12+ 3000 * y13+ 2000 * y14+ 2000 * y21+ 2000 * y22+ 4000 * y23+ 4000 * y24+ 10 * z1+ 5 * z2, GRB.MINIMIZE)

model.optimize()

print("ç·šå½¢è¨ˆç”»æ³•ã®è§£ç­”")
print((model.ObjVal)/10000, "å„„å††")
for v in model.getVars():
    print(v.VarName, v.X)
```
